```python
from datetime import datetime, timedelta, timezone
import pandas as pd
import matplotlib.pyplot as plt
from fbprophet import Prophet
import schedule
import time
import csv
```

```python
data = pd.read_csv('stock_data/data/4661.csv', usecols=[1, 5])
data.columns = ['ds', 'y']
```

```python
model = Prophet(daily_seasonality=True)
#Prophetのモデル作成(daily_seasonality=True)

# 非線形トレンドの指定
#model = Prophet(daily_seasonality=True, growth='logistic')

model.fit(data)
```

```python
# 値動きを予測する関数
future_data = model.make_future_dataframe(periods=365, freq = 'd') #periods=予測する日数
forecast_data = model.predict(future_data)
```

```python
#予測結果をプロット
fig = model.plot(forecast_data)
model.plot_components(forecast_data)
```
